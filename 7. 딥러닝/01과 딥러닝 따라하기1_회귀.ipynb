# 1.환경준비

* 라이브러리 Import

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import *
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

# 2.Regression : Advertising

## (1) 데이터 전처리

### 1) 데이터 준비

```
# 각 매체에 따른 광고비와 매출액 데이터
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/advertising.csv'
adv = pd.read_csv(path)
adv.head()
```

```
target = 'Sales'
x = adv.drop(target, axis=1)
y = adv.loc[:, target]
```

### 2) 가변수화

### 3) 데이터분할

```
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=.2, random_state = 20)
```

## (2) ML 연습 : 선형회귀

```
# 선형회귀 알고리즘을 불러 옵시다.
from sklearn.linear_model import LinearRegression
```

### 1) 모델 선언

```
model = LinearRegression()
```

### 2) 학습

```
model.fit(x_train, y_train)
```

### 3) 예측

```
y_pred = model.predict(x_val)
```

### 4) 검증
만든 모델은 얼마나 정확한지 검증해 봅시다.



```
print(f'RMSE  : {mean_squared_error(y_val, y_pred, squared = False)}')
print(f'MAE   : {mean_absolute_error(y_val, y_pred)}') # 평균 오차
print(f'MAPE  : {mean_absolute_percentage_error(y_val, y_pred)}') # 평균 오차율
# 1- MAPE = 정확도
```

```
# MSE에서 제곱의 효과:

# - 오차가 양수가 되며
# - 0<오차<1은 더 작아지고
# - 1<오차는 더 커짐

# 따라서 에러가 클수록 더 커지는 효과가 있음
# 단, 스케일이 맞지 않아 설명하기 힘듦 → 루트 씌워 RMSE로 사용함
```

## (3) 딥러닝 모델링
* 필요한 함수들 불러오기
* 모델 선언
* 학습
* 예측
* 성능 검증

### 1) 전처리 : Scaling

```
scaler = MinMaxScaler() # 모든 값을 0~1사이의 값으로 Normalization(정규화)
# StandardScaler(표준화)는 평균이 0, 표준편차는 1이 되게끔 바꿔줌 -> 이상치가 심할 때 주로 사용
# But! DL에선 MinMax 주로 사용

x_train = scaler.fit_transform(x_train)
x_val = scaler.transform(x_val)
```

### 2) 필요한 함수들 불러오기

```
from keras.models import Sequential
from keras.layers import Dense
from keras.backend import clear_session
```

### 3) 모델 선언

```
nfeatures = x_train.shape[1] # num of columns(features)
nfeatures
```

```
# 메모리 정리 (옵션)
clear_session()

# 모델 구조 설계 - Sequential 타입 모델 선언
model = Sequential( Dense(1, input_shape = (nfeatures,))  ) # 1은 output

# 모델요약
model.summary()
```

```
# 컴파일
model.compile(optimizer='adam', loss='mse')
```

### 4) 학습

```
model.fit(x_train, y_train)
```

### 5) 예측

```
pred = model.predict(x_val)
```

### 6) 검증
만든 모델은 얼마나 정확한지 검증해 봅시다.



```
print(f'RMSE  : {mean_squared_error(y_val, pred, squared=False)}')
print(f'MAE   : {mean_absolute_error(y_val, pred)}')
print(f'MAPE  : {mean_absolute_percentage_error(y_val, pred)}')
```

# 3.Regression : Carseat

## (1) 데이터 전처리
* 데이터 준비
* 가변수화
* 스케일링(필요하다면)
* 데이터 분할

### 1) 데이터 준비

* 카시트 판매량 데이터

|	변수명	|	설명	|	구분	|
|	----	|	----	|	----	|
|	**Sales** 	|	 **각 지역 판매액(단위 : 1000개)**	|	**Target**	|
|	CompPrice 	|	지역별 경쟁사 판매가격(달러)	|	feature	|
|	Income 	|	가구당 평균 소득액(1000달러)	|	feature	|
|	Advertising 	|	 각 지역, 회사의 광고 예산(1000달러)	|	feature	|
|	Population 	|	 지역 인구수(단위 : 1000명)	|	feature	|
|	Price 	|	 자사 지역별 판매가격(달러)	|	feature	|
|	ShelveLoc 	|	 진열상태(범주 : Bad, Medium, Good)	|	feature	|
|	Age 	|	 지역 인구의 평균 연령	|	feature	|
|	Education 	|	 교육수준(범주 : 10~18)	|	feature	|
|	Urban 	|	 매장이 도심에 있는지 여부(범주 : Yes, No)	|	feature	|
|	US 	|	 매장이 미국에 있는지 여부(범주 : Yes, No)	|	feature	|


* 데이터 경로 : https://raw.githubusercontent.com/DA4BAM/dataset/master/Carseats.csv

```
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/Carseats.csv'
carseat = pd.read_csv(path)
carseat.head()
```

```
target = 'Sales'
x = carseat.drop(target, axis=1)
y = carseat.loc[:, target]
```

### 2) 가변수화

```
cat_cols = ['ShelveLoc', 'Education', 'US', 'Urban']
x = pd.get_dummies(x, columns = cat_cols, drop_first = True)
```

### 3) 데이터분할

```
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=.2, random_state = 20)
```

### 4) Scaling

```
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_val = scaler.transform(x_val)
```

## (2) 모델링
* 필요한 함수들 불러오기
* 모델 선언
* 학습
* 예측
* 성능 검증

```
from keras.models import Sequential
from keras.layers import Dense
from keras.backend import clear_session
```

### 1) 모델 선언

```
nfeatures = x_train.shape[1] # num of columns
nfeatures
```

```
# 메모리 정리(필수는 아님!)
clear_session()

# Sequential 타입 모델 선언
model = Sequential( Dense(1, input_shape= (nfeatures, )) )

# 모델요약
model.summary()
```

```
# 컴파일
model.compile(optimizer = 'adam', loss= 'mse')
```

### 2) 학습

```
model.fit(x_train, y_train)
```

### 3) 예측

```
pred = model.predict(x_val)
```

### 4) 검증
만든 모델은 얼마나 정확한지 검증해 봅시다.



```
print(f'RMSE  : {mean_squared_error(y_val, pred, squared=False)}')
print(f'MAE   : {mean_absolute_error(y_val, pred)}')
```
