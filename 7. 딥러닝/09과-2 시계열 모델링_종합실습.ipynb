```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import *
from sklearn.preprocessing import MinMaxScaler

from keras.models import Sequential
from keras.layers import Dense, Flatten, SimpleRNN, LSTM, GRU
from keras.backend import clear_session
from keras.optimizers import Adam
```

### (2) 필요한 함수 생성

* 학습곡선 함수

```python
# 학습곡선 함수
def dl_history_plot(history):
    plt.figure(figsize=(10,6))
    plt.plot(history['loss'], label='train_err', marker = '.')
    plt.plot(history['val_loss'], label='val_err', marker = '.')

    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid()
    plt.show()
```

* 데이터 2 --> 3차원 변환

```python
# 시계열 데이터 전처리 2차원 --> 3차원으로 변환
def temporalize(x, y, timesteps):
    nfeature = x.shape[1]
    output_x = []
    output_y = []
    for i in range(len(x) - timesteps + 1):
        t = []
        for j in range(timesteps):
            t.append(x[[(i + j)], :])
        output_x.append(t)
        output_y.append(y[i + timesteps - 1])
    return np.array(output_x).reshape(-1,timesteps, nfeature), np.array(output_y)
```

## 2.데이터 준비

### (1) y 만들기
* 2시간 이후의 수요량을 예측해야 합니다.

```python
data['y'] = data['Count'].shift(-2)
data.head()
```

```python
# 2칸을 앞당겼기 때문에 하위 2행의 y값에 NaN이 표시되어 있습니다.
data.tail()
```

```python
# 하위 2행은 삭제합니다.
# 하위 2행 제외하고 다시 붓기
data = data[:-2]
data.tail()
```

### (2) Feature Engineering

```python
# 요일
data['WeekDay'] = data['DateTime'].dt.dayofweek

# 주말여부
data['WeekEnd'] = np.where(data['WeekDay']>4, 1,0)

# hour
data['Hour'] = data['DateTime'].dt.hour
data.head()
```

### (3) x, y 분리

```python
target = 'y'
x = data.drop(['DateTime', target], axis = 1)
y = data.loc[:,target]
```

### (4) 가변수화

```python
cat_cols = ['Seasons','Holiday','FuncDay', 'WeekDay','Hour']
x = pd.get_dummies(x, columns = cat_cols, drop_first = True)
```

### (5) 스케일링

* y 값이 클때는 y도 스케일링 필요

```python
# x 스케일링
scaler_x = MinMaxScaler()
x_s = scaler_x.fit_transform(x)
```

```python
# y는 1차원이므로 수동으로.
y_max, y_min = y.max(), y.min()
y_s = (y - y_min) / (y_max - y_min)
```

### (6) 3차원 구조 만들기

```python
timestep = 24
x2, y2 = temporalize(x_s, y_s, timestep)
x2.shape, y2.shape
```

### (7) 데이터 분할

```python
x_train, x_val, y_train, y_val = train_test_split(x2, y2, test_size= 24*14, shuffle = False)
```

```python
x_train.shape, y_train.shape
```

```python
x_val.shape, y_val.shape
```

## 3.모델링1



### (1) 입력 구조(shape)
* 분석 단위 : 2차원 ( timesteps, nfeatures)

```python
timesteps = x_train.shape[1]
nfeatures = x_train.shape[2]
print(timesteps, nfeatures)
```

### (2) 모델 구조 설계

```python
clear_session()

model = Sequential([
    LSTM(12, input_shape=(timesteps, nfeatures), return_sequences=True),
    LSTM(6, return_sequences=True),
    Flatten(),
    Dense(6, activation='relu'),
    Dense(1)
])

model.summary()
```

### (3) 컴파일 및 학습

```python
model.compile(optimizer=Adam(learning_rate=0.01), loss='mse')
history = model.fit(x_train, y_train, epochs=50, validation_split=.2, verbose= 1).history
```

```python
dl_history_plot(history)
```

### (4) 예측 및 평가

```python
pred = model.predict(x_val)
```

```python
 # 스케일링 했던 y데이터 돌려놓기
 y_val = y_val * (y_max - y_min) + y_min
 pred = pred * (y_max - y_min) + y_min
```

```python
print('mse: ', mean_squared_error(y_val, pred, squared=False))
print('mae: ', mean_absolute_error(y_val, pred))
```

```python
plt.figure(figsize = (20,6))
plt.plot(y_val, label = 'actual')
plt.plot(pred, label = 'predicted')
plt.legend()
plt.grid()
plt.show()
```
