# 1.환경 준비

- 기본 라이브러리와 대상 데이터를 가져와 이후 과정을 준비합니다.

```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action='ignore')
%config InlineBackend.figure_format='retina'
```

```python
# 데이터 읽어오기
path = 'https://raw.githubusercontent.com/jangrae/csv/master/titanic_train.csv'
data = pd.read_csv(path)
```

# 2.데이터 이해

- 분석할 데이터를 충분히 이해할 수 있도록 다양한 탐색 과정을 수행합니다.

# 3.데이터 준비

- 전처리 과정을 통해 머신러닝 알고리즘에 사용할 수 있는 형태의 데이터를 준비합니다.

**1) 변수 제거**

- 분석에 의미가 없다고 판단되는 변수는 제거합니다.

```python
# 제거 대상: PassengerId, Name, Ticket, Cabin
drop_cols = ['PassengerId', 'Name', 'Ticket', 'Cabin']

# 변수 제거
data.drop(drop_cols, axis=1, inplace=True)

# 확인
data.head()
```

**2) 결측치 처리**

- 결측치가 있으면 제거하거나 적절한 값으로 채웁니다.

```python
# Age 결측치를 중앙값으로 채우기
age_median = data['Age'].median()
data['Age'].fillna(age_median, inplace=True)
```

```python
# Embarked 최빈값으로 채우기
emb_freq = data['Embarked'].mode()[0]
data['Embarked'].fillna(emb_freq, inplace=True)
```

**3) x, y 분리**

- 우선 target 변수를 명확히 지정합니다.
- target을 제외한 나머지 변수들 데이터는 x로 선언합니다.
- target 변수 데이터는 y로 선언합니다. 
- 이 결과로 만들어진 x는 데이터프레임, y는 시리즈가 됩니다.
- 이후 모든 작업은 x, y를 대상으로 진행합니다.

```python
# target 확인
target = 'Survived'

# 데이터 분리
x = data.drop(target, axis=1)
y = data.loc[:, target]
```

**4) 가변수화**

- 범주형 변수에 대한 가변수화를 진행합니다.

```python
# 가변수화 대상: Pclass, Sex, Embarked
dumm_cols = ['Pclass', 'Sex', 'Embarked']

# 가변수화
x = pd.get_dummies(x, columns=dumm_cols, drop_first=True, dtype=int)

# 확인
x.head()
```

**5) 학습용, 평가용 데이터 분리**

- 학습용, 평가용 데이터를 적절한 비율로 분리합니다.
- 반복 실행 시 동일한 결과를 얻기 위해 random_state 옵션을 지정합니다.

```python
# 모듈 불러오기
from sklearn.model_selection import train_test_split

# 7:3으로 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

# 4.모델링

- 본격적으로 모델을 **선언**하고 **학습**하고 **평가**하는 과정을 진행합니다.

```python
# 1단계: 불러오기
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# 2단계: 선언하기
model = DecisionTreeClassifier(random_state=1, max_depth=4) # max_depth로 가지치기
```

```python
# 3단계: 학습하기
model.fit(x_train, y_train)
```

```python
# 4단계: 예측하기
y_pred = model.predict(x_test)
```

```python
# 5단계 평가하기
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

Decision Tree가 갖고 있는 랜덤한 특성
- max_feature: 최선의 질문을 고민하기 위해 랜덤하게 몇개의 변수를 고려할 것인가? default = None(전부 다 씀)
- 만약 서로다른 두 변수로 질문했을 때 불순도가 똑같이 줄어들었을 때, 의사결정트리는 두 변수중 랜덤하게 채택해서 분기해나감 - 고정하고 싶으면 역시 random_state

**1) 트리 시각화**

Sex와 Pclass, Age...가 생존여부 예측하는 데 중요한 변수임을 예측 가능

**2) 변수 중요도 시각화**

- 변수 중요도룰 시각화해 봅니다.
- feature_importances_ 속성이 변수 중요도입니다.
- 세로 막대 보다는 가로 막대 그래프로 보면 좋습니다.

```python
# 변수 중요도
print(x.columns)
print(model.feature_importances_)
# 대부분 Tree에서 가장 먼저 물어본 변수의 중요도가 더 크다고 나옴 (아닐 경우도 있음. 중요도는 다 더해서 나오므로)
# -> 왜일까? 가변수화 때문.. 범주가 1000개인 변수는 계속 질문에 나오므로 중요도가 모이다 보면 중요도가 크게 나올 수도 있음 -> 따라서 의사결정트리에선 가변수화 하는 것에 주의해야 함
```

```python
# 변수 중요도
plt.figure(figsize=(5, 5))
plt.barh(list(x), model.feature_importances_)
plt.show()
```

- 필요하면 다음과 같이 중요도를 기준으로 정렬해 시각화합니다.

```python
# 데이터프레임 만들기
df = pd.DataFrame()
df['feature'] = x.columns
df['importance'] = model.feature_importances_
df.sort_values(by='importance', ascending=True, inplace=True)

# 시각화
plt.figure(figsize=(5, 5))
plt.barh(df['feature'], df['importance']) # barh는 정렬된 것과 반대로 보여줌 -> 내림차순으로 보려면 ascending=True로 해줘야 함 (seaborn은 그대로 써주면 됨!)
plt.show()
# max_depth에 따라 아예 질문에 나오지조차 않아 중요도가 0이 되는 변수가 생김
# max_depth라고 그냥 자르는 게 아니라 의사결정트리는 max_depth도 고려해서 질문하게 됨
# max_depth가 None일 경우 모든 변수에 대해 질문하므로 중요하지 않은 변수도 중요도를 부여하므로 과적합이 생길수도 있음
# 따라서 초반에 중요하지 않은 변수를 제거하는 게 도움이 되겠지만, 중요하지 않은 변수를 찾는 건 쉽지가 않음

# ※ 변수의 중요도 확인하기
# 변수 하나씩 제거해가며 r2 score확인해보기 -> 만약 특정 변수를 제거한 채 모델링 했을 때 r2 score가 뚝 떨어지면 제거한 변수는 중요한 변수였던 것
# r2 score가 별로 변하지 않는다면 별로 중요하지 않은 변수였다는 것

# 또한, 원래 있던 변수들 사이에 다중공선성 있다면 제거하기=> y빼놓고 각 변수를 target으로 둔 뒤 나머지 x들을 이용해 r2 score 구하기
# r2 score가 엄청 높게 나온다면, 나머지 변수가 그 변수를 설명한다는 의미-> 그 변수는 제외대상이 됨
```
