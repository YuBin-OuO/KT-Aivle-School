```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as spst
```

```python
df = pd.read_csv('sbikedata.csv')
```

### 이변량 분석

```python
sns.barplot(x='precipitation', y='count', data=df)
plt.show()
```

1시간 전 비가 오지 않았을 때(0일 때) 따릉이 대여량이 매우 높음

1시간 전 비가 왔을 때(1일 때) 따릉이 대여량이 매우 낮음

```python
rain = df.loc[df['precipitation'] == 1, 'count']
nrain = df.loc[df['precipitation'] == 0, 'count']

res = spst.ttest_ind(rain ,nrain)
print(f'강우 여부와 대여량의 t통계량 : {res[0]}  \np-value : {res[1]}')
```

### 가설 검정

- t통계량이 -2보다 매우 작고, p-value 또한 매우 작다
> 강수 여부에 따라 대여량은 큰 차이를 가진다
