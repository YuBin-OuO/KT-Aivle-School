```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.graphics.mosaicplot import mosaic
from scipy import stats as spst
import statsmodels.api as sm
import joblib

import warnings
warnings.filterwarnings('ignore')
```

```python
plt.rc('font', family='Malgun')
sns.set(font="Malgun Gothic", rc={"axes.unicode_minus":False}, style='darkgrid')
```

```python
df = pd.read_csv('sbikedata.csv')
```

```python
df2 = df.copy()
```

## 단변량 분석

* 일별 단변량 분석

```python
plt.figure(figsize=(20, 6))
sns.histplot(df2['date'], kde=True)
plt.show()
```

* 일별 분석의 그래프가 예상과는 많은 차이가 있어 월 별 데이터로 범주 축소

* 월별 단변량 분석

```python
df2['date'] = pd.to_datetime(df2['date'])
```

```python
# 'date' 컬럼을 년, 월, 일로 분리하여 새로운 열로 추가
df2['year'] = df2['date'].dt.year
df2['month'] = df2['date'].dt.month
df2['day'] = df2['date'].dt.day
df2['season'] = np.where(df2['month'] < 5, '봄', np.where(df2['month'] < 8, '여름', np.where(df2['month'] < 11, '가을', '겨울')))
```

```python
plt.figure(figsize=(20, 6))
sns.histplot(df2['month'], kde=True)
plt.show()
```

* 일별 분석과는 다르게 특정 패턴이 보이기 시작
* 계절 단위로 범주 축소

* 계절별 단변량 분석

```python
plt.figure(figsize=(20, 6))
sns.histplot(df2['season'], kde=True)
plt.show()
```

* 월별 분석보다 확실한 패턴 확인

* 계절 단위로 확인 했을 때 데이터의 수가 여름과 가을에 주로 분포 되어있는 것으로 확인
