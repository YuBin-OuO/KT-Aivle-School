## **2.결합1 : 붙이기 concat()**

* 데이터를 생성해 봅시다.

```python
df1 = pd.DataFrame({'A':[10,25], 'B':[15,30]})
df2 = pd.DataFrame({'A':[20, 30, 50], 'C':[35,30, 40]})
df2.drop([1], inplace = True) # 인덱스가 1인 행 제거 (drop의 axis default는 0(행)이므로)

display(df1)
display(df2)
```

### **(1) 세로로 합치기 : axis = 0**

* join = 'inner'

```python
pd.concat([df1, df2], axis = 0, join = 'inner') # 따로 할당해주지 않으면 조회
```

* join = 'outer'

```python
pd.concat([df1, df2], axis = 0, join = 'outer')
```

### **(2) 가로로 합치기 : axis = 1**

* join = 'inner'

```python
pd.concat([df1, df2], axis = 1, join = 'inner')
```

* join = 'outer'

```python
pd.concat([df1, df2], axis = 1, join = 'outer')
```

```python
pd.concat([sales1, sales2, sales5, sales11], axis=0, join='inner')
```

[문2] (조금 어려운 문제!)  
일별, 매장별 판매량 합계를 구해 봅시다.  
groupby, pd.concat 을 이용하여 다음과 같은 결과를 얻도록 코드를 작성해봅시다.

* 힌트
    * 1) sales1, 2, 5,11 로 부터 각각 일별 판매량 합계 집계하기
        * 집계시 .agg() 를 이용하여 열이름 변경하거나, 집계후 열이름 변경하기
    * 2) pd.concat 으로 가로로 합치기



```python
s1 = sales1.groupby('Date', as_index=False)['Qty'].sum()
s1.columns = ['Date', 'Qty1']
s2 = sales2.groupby('Date', as_index=False)['Qty'].sum()
s2.columns = ['Date', 'Qty2']
s5 = sales5.groupby('Date', as_index=False)['Qty'].sum()
s5.columns = ['Date', 'Qty5']
s11 = sales11.groupby('Date', as_index=False)['Qty'].sum()
s11.columns = ['Date', 'Qty11']
pd.concat([s1, s2, s5, s11], axis=1, join='inner')
```

## **3.결합 ① : 조인 join(merge)**

* 판다스에서 join은 굉장히 간단합니다.
* 자동으로 key를 잡아줍니다
* default로 inner join을 합니다.

```python
df1 = pd.DataFrame({'A':[1,2], 'B':[15,30], 'C':[20, 25]})
df2 = pd.DataFrame({'A':[2,3], 'D':[20, 35]})

display(df1)
display(df2)
```

### **(1) inner merge**

```python
pd.merge(df1, df2, how = 'inner', on = 'A')
```

```python
pd.merge(df1, df2, how = 'inner') # on을 생략하면 알아서 같은 것 찾아서 merge
```

### **(2) left merge**

```python
pd.merge(df1, df2, how = 'left')
```

### **(3) right merge**

```python
pd.merge(df1, df2, how = 'right')
```

[문1] 일별, 매장별 판매량 합계를 구해 봅시다.  
groupby, pd.merge 을 이용하여 다음과 같은 결과를 얻도록 코드를 작성해봅시다.

* 힌트
    * 1) sales1, 2, 5,11 로 부터 각각 일별 판매량 합계 집계하기
        * 집계시 .agg() 를 이용하여 열이름 변경하거나, !집계후 열이름 변경하기!
    * 2) pd.merge 으로 결합하기



```python
s1 = sales1.groupby('Date', as_index=False)['Qty'].sum()
s1.columns = ['Date', 'Qty1']
s2= sales2.groupby('Date', as_index=False)['Qty'].sum()
s2.columns = ['Date', 'Qty2']
s5 = sales5.groupby('Date', as_index=False)['Qty'].sum()
s5.columns = ['Date', 'Qty5']
s11= sales11.groupby('Date', as_index=False)['Qty'].sum()
s11.columns = ['Date', 'Qty11']
s = pd.merge(s1, s2, how='inner', on = 'Date')
s = pd.merge(s,s5, how='inner', on = 'Date')
s = pd.merge(s,s11, how='inner', on = 'Date')
s
```

[문2] 매장1에 대해서, 일별, 카테고리별 판매량 합계를 구해 봅시다. 



```python
sales1.head(3)
```

```python
products.head(3)
```

```python
q = pd.merge(sales1, products, how='inner', on = 'Product_ID') # how, on은 생략 가능
q = q.groupby(['Date', 'Category'], as_index = False)['Qty'].sum()
q
```

[문3] 도시별, 총판매량 합계를 구하시오.  



```python
t = pd.merge(sales, stores, how='inner', on='Store_ID')
t.groupby('City', as_index = False)['Qty'].sum()
```

### **(4) pivot**
* pivot : 결합은 아니지만, 집계 후 데이터프레임 구조를 변형해서 조회하는데 종종 사용 됩니다.

* 단계
    * 1) groupby
    * 2) pivot

```python
# 1) 매장1의 일별 카테고리별 판매량을 집계

temp = pd.merge(sales1, products)
temp2 = temp.groupby(['Date', 'Category'], as_index = False)['Qty'].sum()
temp2
```

```python
# 2) pivot
temp3 = temp2.pivot(index = 'Category', columns = 'Date', values = 'Qty')
temp3
```

```python
# sns.heatmap으로 시각화 할 수도 있습니다!

plt.figure(figsize = (20, 6))
sns.heatmap(temp3)
plt.show()
```

## **4.복습문제**

```python
# 라이브러리 불러오기
import pandas as pd
```

```python
sales = pd.read_csv("https://raw.githubusercontent.com/DA4BAM/dataset/master/ts_sales_simple.csv")
products = pd.read_csv("https://raw.githubusercontent.com/DA4BAM/dataset/master/ts_product_master.csv")
stores = pd.read_csv("https://raw.githubusercontent.com/DA4BAM/dataset/master/ts_store_master.csv")
```

1) sales에 매출액(Amt) 변수를 추가하시오.
* Amt = Qty * Price

```python
t = pd.merge(sales, products, how='inner', on='Product_ID')
t['Amt'] = t['Qty']*t['Price']
t
```

2) City별 매출액

```python
t2 = pd.merge(t, stores, how='inner', on='Store_ID')
t2
```

```python
t2.groupby('City', as_index=False)['Amt'].sum()
```

3) City별 카테고리별 매출액

```python
t2.groupby(['City', 'Category'], as_index=False)['Amt'].sum()
t2
```

4) 매출액이 가장 높은 주(state) top 3

```python
t2.groupby('State', as_index=False)['Amt'].sum().sort_values(by = 'Amt',ascending=False)[:3]
```

추가 문제

* 2013년 1월 세째 주 각 도시별 subCategory별 금액 비교
    1. 2013년 1월 세째 주
        * sales['Date']를 날짜 타입으로 변환
        * week 추가
        * sales에서 week가 3인 데이터 뽑기
    2. 각 도시별 subCategory별
        * 1의 결과에서...
        * stores와 inner merge
        * products와 inner merge
    3. 도시별 subCategory별 금액
        * Qty * Price로 Amt 추가
        * groupby
    4. 좀더 보기 좋게! pivot + heatmap
        * pivot
        * heatmap

```python
#1.
sales['Date'] = pd.to_datetime(sales['Date'])
sales['week'] = sales['Date'].dt.isocalendar().week
temp = sales.loc[sales['week']==3]
```

```python
#2.
temp2 = pd.merge(temp, stores)
temp2 = pd.merge(temp2, products)
temp2
```

```python
#3.
temp2['Amt'] = temp2['Qty']*temp2['Price']
temp3 = temp2.groupby(['City', 'SubCategory'], as_index = False)['Amt'].sum()
temp3
```

```python
#4.
piv = temp3.pivot(index = 'City', columns = 'SubCategory', values = 'Amt')
piv
```

```python
plt.figure(figsize = (20, 6))
sns.heatmap(piv)
plt.show()
```
