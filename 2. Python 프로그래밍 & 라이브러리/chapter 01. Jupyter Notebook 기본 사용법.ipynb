아래 실행 단축키 3가지를 다 눌러보고 차이점을 적어 봅시다.

* Ctrl + Enter : 해당 코드 셀에 머물면서 반복 실행 가능
* Shift + Enter : 밑에 빈 코드 셀 만들어지지 않으면서 커서만 아래로 이동
