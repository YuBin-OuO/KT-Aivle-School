### 2) 모델 구조만 빌려와서 사용해보기

* 모델 구조 선택하기

```python
model_scratch = YOLO(model='yolov8n.yaml', task='detect')
```

* 모델 학습하기

```python
model_scratch.train(data='/content/Digits-3/data.yaml',
                    epochs=100,
                    patience=5,
                    seed=2024,
                    pretrained=False,
                    )
```

* 예측해보기

```python
image_path = 'https://www.telegraph.co.uk/content/dam/football/2019/08/12/TELEMMGLPICT000206209364_trans_NvBQzQNjv4BqK3Ytq28vYzV8vgytz3tt20cdhPuOVqLHI0GKTHeusDU.jpeg'
```

```python
model_scratch.predict(source=image_path, stream=False, save=True)
```

### 3) 모델 구조에 사전 학습 가중치도 가져와서 사용해보기

* 모델 구조 및 사전 학습 가중치 선택하기

```python
model_transfer = YOLO(model='yolov8n.pt', task='detect')
```

* 모델 학습하기

```python
model_transfer.train(data='/content/futbol-players-7/data.yaml',
                    epochs=100,
                    patience=5,
                    seed=2024,
                    pretrained=True,
                    )
```

* 예측해보기

```python
model_transfer.predict(source=image_path, save=True,
                       conf=0.05, iou=0.95)
```
