```python
from ultralytics import YOLO
```

```python
model = YOLO() # task=None이면 기본적으로 detect(object detection)
# model = COCO 데이터로 학습하고, 가장 경량화된 모델 → YOLOv8n (nano)을 pytorch로 저장한 것을 가져옴
```

```python
model.train() # data=coco.yaml (깃헙에 올려져 있는 파일을 가져와 학습에 사용)
```

```python
model.val()
```

```python
model.predict(save = True, save_txt=True) # 입력으로 들어온 파일 형태와 txt 파일 형태로 학습 마친 이미지 저장
```
