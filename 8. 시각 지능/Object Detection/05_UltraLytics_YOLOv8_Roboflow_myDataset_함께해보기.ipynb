### 1) 모델 구조만 빌려와서 사용해보기

* 모델 구조 선택하기

```python
model = YOLO(model='yolov8n.yaml', task='detect')
```

* 모델 학습하기

```python
model.train(data='/content/Chiikawa-1/data.yaml',
            epochs=100,
            patience=5,
            pretrained=False,
            verbose=True,
            seed=2024,
            )
```

* 예측해보기

```python
results = model.predict(source='/content/Chiikawa-1/test/images',
                        conf=0.3,
                        iou=0.5,
                        save=True,
                        line_width=2)
```
