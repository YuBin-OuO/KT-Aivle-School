### Gmarket
- 베스트 상품 200개 데이터 수집
- 상품의 이미지 200개 다운로드

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

#### 1. URL 찾기

```python
url = 'https://www.gmarket.co.kr/n/best'
```

#### 2. request > response : str(html)

```python
response = requests.get(url)
response
```

```python
response.text[:200]
```

#### 3. bs > DataFrame

```python
dom = BeautifulSoup(response.text, 'html.parser')
```

```python
# select 200 items
elements = dom.select('#gBestWrap > div.best-list > ul > li')
len(elements)
```

```python
element = elements[0]
element
```

```python
data = {
    'title': element.select_one('.itemname').text,
    'link': element.select_one('.itemname').get('href'),
    'img': 'https:' + element.select_one('.image__lazy').get('src'),
    's_price': element.select_one('.s-price').text.split(' ')[0][3:-1].replace(',','')
}
data
```

```python
items = []
for element in elements:
    data = {
    'title': element.select_one('.itemname').text,
    'link': element.select_one('.itemname').get('href'),
    'img': 'https:' + element.select_one('.image__lazy').get('src'),
    's_price': element.select_one('.s-price').text.split(' ')[0][3:-1].replace(',','')
    }
    items.append(data)
```

```python
items[:2]
```

```python
df = pd.DataFrame(items)
df[:2]
```

```python
df.to_excel('g_items.xlsx', index=False)
```
