## Web Crawling
- 웹 페이지에서 데이터를 수집하는 방법에 대해서 학습

### 웹크롤링 방법

#### 웹페이지의 종류
- 정적인 페이지 : 웹 브라우져에 화면이 한번 뜨면 이벤트에 의한 화면의 변경이 없는 페이지
    → 내용이 바뀌려면 url이 바뀌어야 함
- 동적인 페이지 : 웹 브라우져에 화면이 뜨고 이벤트가 발생하면 서버에서 데이터를 가져와 화면을 변경하는 페이지
    → url이 바뀌지 않는 상태에서 내용이 바뀔 수 있음

#### requests 이용
- 받아오는 문자열에 따라 두가지 방법으로 구분
    - json 문자열로 받아서 파싱하는 방법 : 주로 동적 페이지 크롤링할때 사용
    
    
    - html 문자열로 받아서 파싱하는 방법 : 주로 정적 페이지 크롤링할때 사용
    
    
        
#### selenium 이용
- 브라우져를 직접 열어서 데이터를 받는 방법
→ 기업에서 크롤링 막는 경우 있으므로

### Crawling Naver Stock Data
- 네이버 증권 사이트에서 주가 데이터 수집
- 수집할 데이터 : 일별 kospi, kosdaq 주가, 일별 환율(exchange rate) 데이터
- 데이터 수집 절차
    - 웹서비스 분석 : url
    - 서버에 데이터 요청 : request(url) > response : json(str)
    - 서버에서 받은 데이터 파싱(데이터 형태를 변경) : json(str) > list, dict > DataFrame

```python
import warnings
warnings.filterwarnings('ignore') # 경고 문구 출력 X

import pandas as pd
import requests
```

#### 1. 웹서비스 분석 : url
- pc 웹페이지가 복잡하면 mobile 웹페이지에서 수집하는 것도 좋은 방법!

```python
page_size, page = 60, 1
url = f'https://m.stock.naver.com/api/index/KOSPI/price?pageSize={page_size}&page={page}'
url
```

- Disable Cache 체크 (이미 메모리에 저장되어 있는 데이터 사용하지 않고 서버에서 데이터 계속 가져오겠다는 의미)
- clear network log 후 브라우저에서 더보기 등 클릭



- https 443
- Remote Address: 도메인으로부터 얻어온 서버의 주소
- json 포맷은 리스트 안에 딕셔너리가 있음

#### 2. 서버에 데이터 요청 : request(url) > response : json(str)
- response의 status code가 200이 나오는지 확인
- 403이나 500이 나오면 request가 잘못되거나 web server에서 수집이 안되도록 설정이 된것임
    - header 설정 또는 selenium 사용
- 200이 나오더라도 response 안에 있는 내용을 확인 > 확인하는 방법 : response.text

```python
help(requests.get)
```

```python
response = requests.get(url)
response
```

```python
# response 클래스로 만들어진 객체의 text 확인 (데이터타입 = str)
response.text
```

#### 3. 서버에서 받은 데이터 파싱(데이터 형태를 변경) : json(str) > list, dict > DataFrame

```python
# response 객체의 json 메소드
data = response.json()
data
# 객체 안의 text데이터를 리스트나 딕셔너리로 자동으로 바꿔줌 
```

```python
# 리스트 안에 딕셔너리가 들어있는 데이터는 딕셔너리를 하나의 row 데이터로 만듦
df = pd.DataFrame(data)
columns = ['localTradedAt', 'closePrice']
df = df[columns]
df.tail(2)
```

#### 4. 함수로 만들기

```python
def stock_price(code = 'KOSPI', page_size = 60, page = 1):
    # 1. 웹서비스 분석: url 찾기
    url = f'https://m.stock.naver.com/api/index/{code}/price?pageSize={page_size}&page={page}'
    
    # 2. request(url) → response (json): json(str)
    response = requests.get(url)
    
    # 3. json(str) → list, dict → DataFrame
    data = response.json()
    df = pd.DataFrame(data)
    return df[['localTradedAt', 'closePrice']]
```

```python
stock_price('KOSDAQ').tail(2)
```

#### 5. 원달러 환율 데이터 수집 : 실습

```python
def exchange_rate(code = 'FX_USDKRW', page = 1, page_size = 60):
    # 1. 웹서비스 분석: url 찾기
    url = f'https://m.stock.naver.com/front-api/v1/marketIndex/prices?category=exchange&reutersCode={code}&page={page}&pageSize={page_size}'
    
    # 2. request(url) → response (json): json(str)
    response = requests.get(url)
    
    # 3. json(str) → list, dict → DataFrame
    data = response.json()['result'] # 딕셔너리의 result키 값만 가져오기
    df = pd.DataFrame(data)
    return df[['localTradedAt', 'closePrice']]
```

```python
exchange_rate()
```

#### 6. 시각화

```python
 # 주피터 노트북에서 matplotlib 플롯을 직접 노트북 안에 표시할 수 있도록 함. 이 명령이 없으면 plt.show()를 사용하여 플롯을 별도의 창에 표시해야 함
%matplotlib inline

# inline 표시 위한 figure 포맷
%config InlineBackend.figure_formats = {'png', 'retina'}
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 수집
page_size = 60
kospi_df = stock_price('KOSPI', page_size = page_size)
kosdaq_df = stock_price('KOSDAQ', page_size = page_size)
usd_df = exchange_rate('FX_USDKRW', page_size = page_size)
```

```python
# 데이터 전처리 1: 데이터 타입 변경
print(kospi_df.dtypes)
kospi_df['kospi'] = kospi_df['closePrice'].apply(lambda data: float(data.replace(',', '')))
kospi_df = kospi_df.drop(columns = ['closePrice'])
print(kospi_df.dtypes)
```

```python
kosdaq_df['kosdaq'] = kosdaq_df['closePrice'].apply(lambda data: float(data.replace(',', '')))
kosdaq_df = kosdaq_df.drop(columns = ['closePrice'])
```

```python
usd_df['usd'] = usd_df['closePrice'].apply(lambda data: float(data.replace(',', '')))
usd_df = usd_df.drop(columns = ['closePrice'])
```

```python
# 데이터 전처리 2: 날짜 데이터 맞추기 (merge)
merge_df_1 = pd.merge(kospi_df, kosdaq_df, on='localTradedAt')
merge_df_2 = pd.merge(merge_df_1, usd_df, on='localTradedAt')
merge_df = merge_df_2.copy()
merge_df.tail(2)
```

```python
plt.figure(figsize=(16, 8))
columns = merge_df.columns[1:]
for column in columns:
    plt.plot(merge_df['localTradedAt'], merge_df[column], label=column)
    
xticks_count = 11
plt.xticks(merge_df['localTradedAt'][::int(len(merge_df) // xticks_count) +1])
plt.legend(loc=0) # 위치 알아서
plt.show()
```

#### 7. 데이터 스케일링
- min max scaling


- $z = \frac{x_i - min(x)}{max(x) - min(x)} (0 \leqq z \leqq 1)$


- latex syntax : `https://jjycjnmath.tistory.com/117`

```python
from sklearn.preprocessing import minmax_scale
```

```python
# 시각화
plt.figure(figsize=(20,5))
columns = merge_df.columns[1:]
for column in columns:
    plt.plot(merge_df['localTradedAt'], minmax_scale(merge_df[column]), label=column)

xticks_count = 11
plt.xticks(merge_df['localTradedAt'][::int(len(merge_df) // xticks_count) +1])
plt.legend(loc=0) # 위치 알아서
plt.show()
```

#### 8. 상관관계 분석
- 피어슨 상관계수(Pearson Correlation Coefficient)
- 두 데이터 집합의 상관도를 분석할때 사용되는 지표
- 상관계수의 해석
    - -1에 가까울수록 서로 반대방향으로 움직임
    - 1에 가까울수록 서로 같은방향으로 움직임
    - 0에 가까울수록 두 데이터는 관계가 없음

```python
corr_df = merge_df[merge_df.columns[1:]].corr()
corr_df
```

더 많은 양의 데이터일 경우


+ kospi와 kosdaq은 0.8
+ kosdaq과 usd는 -0.6
+ kospi와 usd는 -0.6 정도

```python
# 결정계수: r-squared
# 1과 가까울수록 강하고, 0과 가까울수록 약한 관계
plt.figure(figsize=(20,5))
sns.heatmap(corr_df**2, annot=True)
plt.show()
```
