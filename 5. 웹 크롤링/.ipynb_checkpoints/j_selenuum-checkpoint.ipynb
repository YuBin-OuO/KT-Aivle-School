```python
driver.find_element(By.CSS_SELECTOR, '.viewScroll-Center').click()
```

```python
text=''
```

```python
while 1:
    # 텍스트 선택
    episode = driver.find_element(By.CSS_SELECTOR, '.text-wrap')

    # 텍스트 데이터 가져오기
    episode = episode.text.replace('조아라에 게시된 모든 작품은 저작권법에 의거 보호받고 있습니다 ※\n※ 저작권자의 승인 없이 작품의 일부, 또는 전부를 복제, 전송, 배포 및 기타의 방법으로 이용할 경우,손해배상 청구를 포함해 강력한 민/형사상 처벌대상이 됩니다. (5년 이하의 징역, 5천만원 이하의 벌금부과)','')
    episode = episode.replace('\n\n\n', '')
    
    print(episode)
    text += episode + '\n\n'
    
    # 딜레이
    time.sleep(random.uniform(2.2, 5))
    
    # 다음화
    driver.find_element(By.CSS_SELECTOR, '[data-testid="joa-viewerBottom-right"]').click()
    
    # 딜레이
    time.sleep(random.uniform(2.2, 5))
```

```python
# 브라우져 종료
driver.quit() # 100mb 넘게 브라우저가 차지하고 있던 메모리 반환. 그냥 x로 끄면 메모리에 남아있으므로
```
