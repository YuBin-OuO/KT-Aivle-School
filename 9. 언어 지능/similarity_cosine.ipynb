```python
from sklearn.feature_extraction.text import TfidfVectorizer # 텍스트에서 단어 뽑아내 벡터 만들기 위함
from sklearn.metrics.pairwise import cosine_similarity # 코사인 함수 이용하기 위함
```

```python
sentence = ('오늘도 폭염이 이어졌는데요, 내일은 반가운 비 소식이 있습니다.',
            '폭염을 피해 놀러왔다가 갑작스런 비로 망연자실하고 있습니다.')

vector = TfidfVectorizer(max_features=100) # 문장을 크기 100의 벡터로 바꾸도록
tfidf_vector = vector.fit_transform(sentence)

print(cosine_similarity(tfidf_vector[0], tfidf_vector[0]))
print(cosine_similarity(tfidf_vector[0], tfidf_vector[1]))
```
