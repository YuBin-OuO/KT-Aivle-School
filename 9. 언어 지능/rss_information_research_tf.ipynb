```python
import feedparser # RSS를 입력받았을 때 이를 쪼개서 <item>...</item> 불러올 수 있게끔 함
from newspaper import Article # 각 tag에 대해 쪼개서 처리 가능
from konlpy.tag import Okt # 한국어를 위한 자연어 처리 라이브러리 (명사 추출 시 사용할 것)
from collections import Counter # 명사가 본문에 몇번 나오는지 개수 세기 위함 (TF 계산용)
from bs4 import BeautifulSoup # 인터넷에서 기사 긁어올 수 있게끔 함 (+ 전처리 - html 태그 제거)
```

https://www.etnews.com/rss/

```python
# 1. 모든 RSS 파일 돌아다니며 기사의 title, link 추출
urls = ['http://rss.etnews.com/Section901.xml',
        'http://rss.etnews.com/Section902.xml',
        'http://rss.etnews.com/Section903.xml',
        'http://rss.etnews.com/Section904.xml']


# RSS 목록의 리스트 안에 존재하는 모든 기사의 title와 link를 리스트로 반환
def crawl_rss(urls):
    array_rss = []
    set_titles = set() # 중복 기사 제거 (기사 제목들의 집합 구성)
    for url in urls:
        print('[Crawl URL]', url)
        parse_rss = feedparser.parse(url) # 현재 url을 파싱한 뒤 결과를 parse_rss에 저장
        for p in parse_rss.entries: # parse_rss에 있는 모든 entreis(기사) 검색
            if p.title not in set_titles: # 중복 기사 제거
                array_rss.append({'title': p.title, 'link': p.link}) # title, link 추출해 리스트에 추가
                set_titles.add(p.title)
            else:
                print('Duplicated Article:', p.title)
    return array_rss
```

```python
list_articles = crawl_rss(urls)
```

```python
# 불용어(不用語: stop words) 제거 (등, 수 등은 불용어 (stop words))
def preprocessing(text):
    text_article = BeautifulSoup(text, 'html5lib').get_text() # html 태그 제거
    return text_article
```

```python
# 2. 리스트의 모든 link 돌아다니면서 본문 text 긁어오기
# 언어는 한글로 기본 지정

def crawl_article(url, language='ko'):
    print('[Crawl Article]', url) # title과 text를 추출할 현재 url 출력
    a = Article(url, language = language)
    a.download()
    a.parse()
    return a.title, preprocessing(a.text) # 기사의 title과 text 출력
```

```python
for article in list_articles:
    _, text = crawl_article(article['link'])
    article['text'] = text
```

```python
# 3. 모든 본문 text를 돌아다니면서 명사(키워드)의 빈도수 추출

def get_keywords(text, num_keywords = 10): # num_keywords는 뽑아낼 키워드 개수
    spliter = Okt() # konlpy에 의해 문장을 형태소별로 쪼개는 기능을 위해 spliter 생성
    nouns = spliter.nouns(text) # text의 명사만 출력해 리스트로 추출
    count = Counter(nouns) # 추출된 명사들의 출현 빈도 추출
    list_keywords = []
    for n, c in count.most_common(num_keywords): # num_keywords개의 최빈 키워드 내림차순으로 추출
        item = {'keyword':n, 'count':c}
        list_keywords.append(item)
    return list_keywords
```

```python
for article in list_articles:
    keywords = get_keywords(article['text'])
    article['keywords'] = keywords
```

```python
# 4. 글(단어/문장)을 입력 받아서 그 검색어를 가지고 있는 기사들을 모두 출력

def search_articles(query, list_keywords): # 질의어가 키워드 리스트에 있으면 빈도수 출력 (없으면 0)
    num_words = 0
    for kw in list_keywords:
        if query == kw['keyword']:
            num_words = kw['count']
    return num_words
```

```python
# 하나만 출력
query = input('질의어 입력: ')

for article in list_articles:
    num_query = search_articles(query, article['keywords'])
    if num_query != 0:
        print('[TF]: %2d'%num_query, end = ' ')
        print('[Title]: ', article['title'], end = ' ')
        print('[URL]: ', article['link'])
```

```python
# 내림차순으로 여러개 출력
query = input('질의어 입력: ')
array_query = []

for article in list_articles:
    num_query = search_articles(query, article['keywords'])
    if num_query != 0:
        array_query.append([num_query, article['title'], article['link'], article['text']])

array_query.sort(key = lambda x: -x[0]) # 내림차순 정렬: -x[0]
# array_query.sort(reverse = True, key = lambda x: x[0]) # 이렇게 써줘도 ㄱㅊ
print(array_query, end='\n\n')

print('*** 검색 결과 ***')
for index_query in array_query:
    print('[Title]: ', index_query[1])
    print('[URL]: ', index_query[2])
    print('[Text(50)]: ', index_query[3][:50])
    print('[Frequency/Similarity]: ', index_query[0])
    print()
```
