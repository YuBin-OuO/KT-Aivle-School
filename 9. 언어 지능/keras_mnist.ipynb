```python
from keras.utils import to_categorical
from keras.datasets import mnist
```

```python
# MNIST 데이터
(train_images, train_labels) ,(test_images, test_labels) = mnist.load_data()
```

```python
print(train_images.shape, train_labels.shape, test_images.shape, test_labels.shape)
```

```python
# 0~1 사이의 값으로 바꿔주고 flatten
train_images = train_images.reshape(train_images.shape[0], 784).astype('float32')/255.0 # 28*28 = 784
test_images = test_images.reshape(test_images.shape[0], 784).astype('float32')/255.0
```

```python
# One-Hot Encoding
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# 딥러닝 1단계: 모델 만들기
model = Sequential()
# model.add(Dense(1, activation = 'relu')) # 압축효과가 생기므로 그 이후에 아무리 노드 개수 늘려준대도 성능이 떨어질 것임
# 따라서, 만약 input 개수가 n개라면, 첫번째 은닉층 노드 개수는 최소한 n개는 되도록 해야 압축이 일어나지 않음
model.add(Dense(256, activation = 'relu')) # 여기서도 최소 784는 줘야 하는 게 많지만, mnist데이터는 흑백이기도 하고 워낙 깔끔한 데이터라 ㄱㅊ
model.add(Dense(256, activation = 'relu'))
model.add(Dense(256, activation = 'relu'))
model.add(Dense(10, activation = 'softmax')) # 다중분류이므로
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics=['accuracy']) # 2차원 이미지 optimizer는 adam 쓰는 게 좋음
```

```python
# 딥러닝 2단계: 학습
model.fit(train_images, train_labels, epochs=5, batch_size=128, verbose=1)
```

```python
# 딥러닝 3단계: 추론/테스트
_, accuracy = model.evaluate(test_images, test_labels) # loss와 accuracy값이 나옴
print('Accuracy: %6.2f%%' %(accuracy*100))
```
