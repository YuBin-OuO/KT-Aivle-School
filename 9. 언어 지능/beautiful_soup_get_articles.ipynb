```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = urlopen('https://www.etnews.com/20240416000269')
soup = BeautifulSoup(url, 'lxml') # lxml: BeautifulSoup에서 HTML 또는 XML 문서를 파싱하는 데 사용되는 라이브러리(parser)

title = soup.find('div', {'class': 'article_title'}).get_text(strip=True) # <div> 태그이면서 class 속성이 article_title인 요소를 찾아 title 변수에 저장
# get_text(strip=True): 태그 제거
contents = soup.find('div', {'class': 'article_txt'}).get_text(strip=True)

print(title)
print(contents)
```

BeautifulSoup로 긁어오면 태그까지 그대로 가져옴 -> get_text로 해줘야 함
